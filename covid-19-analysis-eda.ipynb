## Missing value handling

```python
for column in df.columns:
    print(column, df[column].isna().sum())
```

```python
df.drop(df.columns[df.isnull().mean() > 0.60], axis=1, inplace=True)
```

```python
df.isnull().mean() * 100
```

```python
missing_values = df.isna().sum()
columns_with_missing = missing_values[missing_values > 0]
df[columns_with_missing.index].dtypes
```

```python
int_float_columns = df.select_dtypes(include=['int64', 'float64']).columns
df[int_float_columns] = df[int_float_columns].fillna(df[int_float_columns].mean())
```

```python
object_columns = df.select_dtypes(include=['object']).columns
df[object_columns] = df[object_columns].fillna(df[object_columns].mode().iloc[0])
```

## Visualizing

```python
location_counts = df['continent'].value_counts()
plt.figure(figsize=(10, 6))
location_counts.plot(kind='bar', color='skyblue')
plt.title('Count of Each continent')
plt.xlabel('continent')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
average_cases_per_continent = df.groupby('continent')['total_cases_per_million'].mean().reset_index()
plt.figure(figsize=(10, 6))
plt.bar(average_cases_per_continent['continent'], average_cases_per_continent['total_cases_per_million'], color='navy')
plt.xlabel('Continent')
plt.ylabel('Average total Cases per Million')
plt.title('Average total Cases per Million by Continent')
plt.show()
```

```python
average_cases_per_continent = df.groupby('continent')['total_deaths_per_million'].mean().reset_index()
plt.figure(figsize=(10, 6))
plt.bar(average_cases_per_continent['continent'], average_cases_per_continent['total_deaths_per_million'], color='#13697C')
plt.xlabel('Continent')
plt.ylabel('Average total deaths per Million')
plt.title('Average total deaths per Million by Continent')
plt.show()
```

```python
continents = df['continent'].unique()
fig, axes = plt.subplots(nrows=len(continents), ncols=1, figsize=(10, 6 * len(continents)))

for i, continent in enumerate(continents):
    subset_df = df[df['continent'] == continent]
    if not subset_df.empty:
        location_counts = subset_df['location'].value_counts()
        ax = axes[i]
        location_counts.plot(kind='bar', ax=ax, color='skyblue')
        ax.set_title(f'Count of Locations in {continent}')
        ax.set_xlabel('Location')
        ax.set_ylabel('Count')
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Specify the locations and variables you want to compare
selected_locations = ['Costa Rica', 'El Salvador', 'Georgia', 'Montserrat', 'Nauru', 'British Virgin Islands']
variables = ['extreme_poverty', 'reproduction_rate', 'stringency_index', 'hospital_beds_per_thousand', 'new_people_vaccinated_smoothed_per_hundred', 'diabetes_prevalence']

# Filter the DataFrame for the selected locations and variables
filter_loc = df[df['location'].isin(selected_locations)]
filtered_df = filter_loc[variables + ['location']]
avg_values = filtered_df.groupby('location').mean()

# Set up subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 8))
fig.suptitle('Barplot Comparison of Average Variables Across Different Locations', fontsize=16)

for i, variable in enumerate(variables):
    ax = axes.flatten()[i]
    sns.barplot(x=avg_values.index, y=avg_values[variable], ax=ax)
    ax.set_title(f'{variable.capitalize()}')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')  # Rotate x-axis labels

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])
# fig.legend(labels=selected_locations, loc='upper right', title='Location')
plt.show()
```
